```python
import pandas as pd 
import os 

imagesPath = "./dataset/train"
csvPath = './dataset/train_set_1.csv'

df = pd.read_csv(csvPath)

idlst = df['id'].tolist()

imageslst = os.listdir(imagesPath)

imageslstmain = [x.split(".")[0] for x in imageslst]

#create a new list called finalids which contains id's common in imageslstmain and idlst
finalids = [x for x in imageslstmain if x in idlst]

print(finalids)
```
